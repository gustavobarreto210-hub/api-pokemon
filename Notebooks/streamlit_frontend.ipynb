
## 📦 Bloco 1: Importações e Configuração de Página
Análise das bibliotecas visuais, manipuladores de arquivos e configurações básicas do navegador.


```python
import base64
from pathlib import Path
import requests  # type: ignore
import streamlit as st  # type: ignore

st.set_page_config(
    page_title="Pokédex",
    page_icon="⚡",
    layout="centered"
)

API_URL = "http://127.0.0.1:8000"

BASE_DIR = Path(__file__).parent
IMAGE_PATH = BASE_DIR / "assets" / "pokemon.jpg"
```

### Explicação linha por linha:
* `import base64`: Importa o módulo do Python para codificar binários em texto. Essencial para transformar nossa imagem física de fundo em uma string legível pela tag de estilização CSS.
* `from pathlib import Path`: Utilitário moderno para manipulação de diretórios que resolve problemas de caminhos de arquivos de forma independente do sistema operacional (Windows/Linux).
* `import requests`: Cliente HTTP síncrono para enviar requisições do frontend e capturar as respostas JSON enviadas pela nossa API FastAPI.
* `import streamlit as st`: O framework web que gera todos os inputs, botões e renderizações dinâmicas na página através de comandos simples em Python.
* `# type: ignore`: Avisa o linter ou ferramentas de checagem estática para não gerarem alertas de importação dessas bibliotecas terceiras.
* `st.set_page_config(...)`: Modifica as propriedades fundamentais da janela do navegador.
  * `page_title`: Define o texto da aba como "Pokédex".
  * `page_icon`: Define o ícone visual da aba usando um emoji.
  * `layout="centered"`: Mantém todo o conteúdo centralizado na página, evitando que os elementos fiquem espalhados em telas ultrawide.
* `API_URL = "http://127.0.0.1:8000"`: Define a localização exata do nosso servidor backend local.
* `BASE_DIR = Path(__file__).parent`: Localiza a pasta exata onde este script de interface está salvo.
* `IMAGE_PATH = BASE_DIR / "assets" / "pokemon.jpg"`: Cria a rota exata para carregar a imagem de papel de parede do projeto.


## 🎨 Bloco 2: Lógica de Carregamento de Imagem e Estilização Avançada (Injeção de CSS)
Como o Streamlit não possui temas nativos muito customizáveis por padrão, este bloco constrói uma solução sob medida usando tags HTML `<style>` para criar uma identidade visual temática escura inspirada no universo Pokémon.


```python
def adicionar_background():
    if not IMAGE_PATH.exists():
        st.error(f"Imagem não encontrada em: {IMAGE_PATH}")
        st.stop()

    with open(IMAGE_PATH, "rb") as img:
        img_base64 = base64.b64encode(img.read()).decode()

    st.markdown(
        f"""
        <style>
        .stApp {{
            background:
                linear-gradient(
                    rgba(0,0,0,0.55),
                    rgba(0,0,0,0.55)
                ),
                url("data:image/jpeg;base64,{img_base64}");
            background-size: cover;
            background-position: center;
            background-repeat: no-repeat;
            background-attachment: fixed;
        }}
        h1, h2, h3, h4, h5, h6,
        p, span, label {{
            color: white !important;
        }}
        
        /* CORREÇÃO AQUI: Garante de forma abrangente que a letra fique preta ao digitar na barra */
        .stTextInput input, 
        .stTextInput > div > div > input,
        .stTextInput input:focus {{
            color: #000000 !important; /* Letra preta ao digitar */
            background: rgba(255,255,255,.92) !important;
            border-radius:14px;
            height:55px;
            font-size:20px;
            text-align:center;
            font-weight:bold;
        }}
        
        /* Ajusta o texto de exemplo (placeholder) para um cinza escuro legível */
        .stTextInput input::placeholder {{
            color: #555555 !important;
        }}

        .stButton > button {{
            width:100%;
            height:55px;
            font-size:18px;
            border-radius:14px;
            background:#ffcb05;
            color:#222;
            font-weight:bold;
            transition:.25s;
        }}
        .stButton > button:hover {{
            background:#2a75bb;
            color:white;
            transform:scale(1.02);
        }}
        .pokemon-card {{
            background:rgba(15,15,15,.72);
            backdrop-filter:blur(18px);
            border:3px solid #ffcb05;
            border-radius:25px;
            padding:35px;
            margin-top:35px;
            box-shadow:0 0 40px rgba(255,203,5,.45);
        }}
        .badge {{
            display:inline-block;
            background:#ffcb05;
            color:black;
            font-weight:bold;
            padding:6px 14px;
            margin:4px;
            border-radius:999px;
        }}
        .info-box {{
            background:rgba(255,255,255,.08);
            border-radius:15px;
            padding:15px;
            text-align:center;
            margin-bottom:15px;
        }}
        </style>
        """,
        unsafe_allow_html=True
    )

adicionar_background()
```

### Explicação linha por linha:
* `def adicionar_background():`: Cria uma função responsável por gerenciar toda a folha de estilo customizada da aplicação.
* `if not IMAGE_PATH.exists():`: Cláusula de segurança. Valida se a imagem física de fundo está na pasta correta.
* `st.error(...)` e `st.stop()`: Exibe um alerta visual vermelho de erro crítico e encerra sumariamente a execução se a imagem faltar, impedindo crashs do sistema.
* `with open(IMAGE_PATH, "rb") as img:`: Abre o arquivo de imagem em modo de leitura binária.
* `img_base64 = base64.b64encode(img.read()).decode()`: Transforma os dados binários da imagem em uma string Base64. Isso permite incluir a imagem diretamente no código fonte do CSS sem precisar de uma URL de internet.
* `st.markdown(f"...", unsafe_allow_html=True)`: Envia o bloco CSS diretamente para o navegador do usuário. O `unsafe_allow_html=True` é obrigatório para o Streamlit aceitar tags HTML/CSS.
  * `.stApp`: Modifica o container raiz. Aplica um `linear-gradient` escuro (55% de opacidade) sobreposto à nossa imagem em Base64 para garantir a legibilidade das letras. Define as regras de posicionamento como fixas e redimensionáveis (`cover`).
  * `h1, h2..., label`: Força que todas as letras, cabeçalhos e rótulos nativos do Streamlit fiquem na cor branca utilizando o modificador `!important`.
  * `.stTextInput input...`: CSS avançado que corrige o problema crônico de fontes invisíveis no Streamlit. Garante um fundo branco opaco e força a cor da fonte digitada para preto (`#000000`) em todos os estados da caixa (inclusive focada).
  * `::placeholder`: Estiliza o texto de instrução interna da caixa de texto com uma cor cinza escura bem visível.
  * `.stButton > button`: Altera o botão de busca. Define largura total, altura confortável, bordas arredondadas e as cores da marca (fundo amarelo com letra preta).
  * `:hover`: Implementa um efeito visual interativo. Quando o usuário passa o mouse, o botão muda suavemente de cor para azul, a letra vira branca e ele expande levemente (`scale(1.02)`).
  * `.pokemon-card`, `.badge`, `.info-box`: Define classes utilitárias para estruturar caixas transparentes com efeito de vidro fosco (`backdrop-filter: blur`), pequenas pílulas amarelas para as habilidades e cartões de informação técnica.
* `adicionar_background()`: Invoca e aplica a função que acabamos de estudar.


## ⌨️ Bloco 3: Título da Página e Campo de Entrada Textual
Neste ponto, renderizamos o cabeçalho central e preparamos a variável de estado que receberá os comandos de busca do usuário.


```python
st.markdown(
    """
    <h1 style='text-align:center;font-size:55px;'>
        ⚡ Pokédex
    </h1>
    <h4 style='text-align:center;color:white;'>
        Pesquise qualquer Pokémon
    </h4>
    """,
    unsafe_allow_html=True
)

nome = st.text_input(
    "Nome do Pokémon",
    placeholder="Ex: Pikachu"
)

pokemon = None
```

### Explicação linha por linha:
* `st.markdown("<h1 style=...", unsafe_allow_html=True)`: Insere um título principal e um subtítulo no topo da Pokédex, centralizando-os e definindo o tamanho da fonte diretamente via estilo HTML em linha.
* `nome = st.text_input(...)`: Cria uma caixa de texto interativa na tela. O texto que o usuário digita é automaticamente armazenado na variável dinâmica `nome`. O parâmetro `placeholder` exibe a dica esmaecida "Ex: Pikachu".
* `pokemon = None`: Cria a variável controladora vazia. É fundamental declará-la aqui como nula para evitar que o código de exibição que está mais abaixo dê um erro de "variável não definida" antes do botão de busca ser acionado.


## 📡 Bloco 4: Clique do Botão, Consumo da API e Tratamento de Exceções de Rede
Este bloco lida com a reatividade do Streamlit. Quando o botão é pressionado, o app cria uma ponte de comunicação com nossa API FastAPI e trata os possíveis cenários de erro do servidor de forma amigável.


```python
if st.button("Buscar Pokémon"):
    if not nome.strip():
        st.warning("Digite o nome de um Pokémon.")
    else:
        with st.spinner("Buscando na Pokédex..."):
            try:
                resposta = requests.get(
                    f"{API_URL}/pokemon/{nome.lower().strip()}",
                    timeout=10
                )
                
                if resposta.status_code == 404:
                    st.error("Pokémon não encontrado na base de dados.")
                elif resposta.status_code != 200:
                    st.error(f"O servidor retornou um erro inesperado (Status {resposta.status_code}).")
                else:
                    pokemon = resposta.json()

            except requests.Timeout:
                st.error("O servidor demorou muito para responder. Tente novamente.")
            except requests.ConnectionError:
                st.error("Não foi possível conectar à API. Certifique-se de que o backend FastAPI está rodando.")
            except requests.RequestException:
                st.error("Erro crítico de comunicação com a API.")
```

### Explicação linha por linha:
* `if st.button("Buscar Pokémon"):`: Condicional reativa do Streamlit. Toda a lógica abaixo só é executada se o usuário clicar no botão amarelo na interface.
* `if not nome.strip():`: Valida se o usuário preencheu o campo apenas com espaços vazios ou se não digitou nada. O `st.warning` joga um aviso amarelo na tela caso positivo.
* `with st.spinner("Buscando..."):`: Cria um efeito visual flutuante de carregamento animado na tela. O efeito some automaticamente assim que as linhas de código recuadas dentro dele finalizarem.
* `resposta = requests.get(...)`: Envia a requisição HTTP GET para a nossa API FastAPI. O nome digitado é tratado com `.lower().strip()` para garantir o envio correto de letras minúsculas sem espaços. O `timeout=10` encerra a tentativa se o backend travar por mais de 10 segundos.
* `if resposta.status_code == 404:`: Se nossa API retornar o erro 404, o frontend avisa ao usuário através do componente vermelho `st.error` que aquele Pokémon não existe.
* `elif resposta.status_code != 200:`: Captura falhas de servidor (erros 500, 502, 503) e cospe na tela o número do status do problema para auditoria.
* `else: pokemon = resposta.json()`: Cenário perfeito. A requisição foi bem-sucedida (status 200) e salvamos os dados limpos convertidos em um dicionário Python na variável `pokemon`.
* `except requests.Timeout:`: Captura se o nosso backend estourou o limite de tempo estabelecido.
* `except requests.ConnectionError:`: Um erro de infraestrutura muito comum. Ocorre se o desenvolvedor esquecer de rodar o servidor FastAPI no terminal antes de subir o Streamlit. Mostra um aviso cirúrgico instruindo a ligar o backend.
* `except requests.RequestException:`: Um bloco abrangente para capturar qualquer outro tipo de erro inesperado durante a conexão HTTP.


## 🖼️ Bloco 5: Exibição Gráfica Dinâmica e Estrutura dos Dados Filtrados
Se a busca foi realizada com sucesso e a variável `pokemon` foi povoada com os dados do JSON, a interface monta uma ficha técnica organizada por colunas com o nome, imagem, atributos fundamentais e as habilidades do monstrinho.


```python
if pokemon:
    
    st.markdown(
        f"""
        <h1 style="text-align:center; color:#ffcb05 !important;">
             {pokemon['nome'].capitalize()}
        </h1>
        """,
        unsafe_allow_html=True
    )

    c1, c2, c3 = st.columns([1, 2, 1])
    with c2:
        if pokemon.get("imagem"):
            st.image(pokemon["imagem"], width=260)
        else:
            st.info("Imagem não disponível")

    st.markdown("---")

    col1, col2 = st.columns(2)
    with col1:
        st.markdown(
            f"""
            <div class='info-box'>
                <h4>ID</h4>
                <h2>#{pokemon['id']}</h2>
            </div>
            <div class='info-box'>
                <h4>Altura</h4>
                <h2>{pokemon['altura']} m</h2>
            </div>
            """,
            unsafe_allow_html=True
        )

    with col2:
        st.markdown(
            f"""
            <div class='info-box'>
                <h4>Peso</h4>
                <h2>{pokemon['peso']} kg</h2>
            </div>
            <div class='info-box'>
                <h4>Tipo</h4>
                <h2>{', '.join(pokemon['tipos']).title()}</h2>
            </div>
            """,
            unsafe_allow_html=True
        )

    st.markdown("### ✨ Habilidades")
    habilidades_html = ""
    for habilidade in pokemon.get("habilidades", []):
        habilidades_html += f"<span class='badge'>{habilidade.title()}</span>"
    
    st.markdown(habilidades_html, unsafe_allow_html=True)
    
    st.markdown("</div>", unsafe_allow_html=True)
```

### Explicação linha por linha:
* `if pokemon:`: Garante que toda a renderização abaixo permaneça totalmente oculta na tela até que um Pokémon válido seja pesquisado e encontrado.
* `{pokemon['nome'].capitalize()}`: Captura a chave `'nome'` do dicionário trazido pela API e aplica a função `.capitalize()` do Python para deixar apenas a primeira letra em maiúscula (ex: "Pikachu"), renderizando o texto em destaque amarelo.
* `c1, c2, c3 = st.columns([1, 2, 1])`: Cria três colunas invisíveis lado a lado. O array numérico define as proporções: a coluna do meio (`c2`) terá o dobro do tamanho das colunas laterais.
* `with c2:`: Entra no escopo da coluna do meio.
* `st.image(pokemon["imagem"], width=260)`: Se houver uma string de imagem válida trazida pelo JSON, renderiza a arte oficial centralizada na tela com largura fixa de 260px. Caso contrário, exibe um informativo azul de imagem indisponível.
* `st.markdown("---")`: Desenha uma linha horizontal elegante para separar a imagem dos dados técnicos.
* `col1, col2 = st.columns(2)`: Cria duas colunas idênticas e simétricas na tela para distribuir os dados.
* `with col1:`: Na coluna da esquerda, monta blocos HTML usando a classe personalizada `.info-box` definida no CSS inicial para exibir de forma polida e limpa o **ID** (ex: #25) e a **Altura** do Pokémon já convertida para metros.
* `with col2:`: Na coluna da direita, faz exatamente o mesmo processo para exibir o **Peso** em quilogramas (kg) e o **Tipo**.
* `', '.join(pokemon['tipos']).title()`: Trata a lista de tipos. Se o Pokémon possuir dois tipos (ex: `['grass', 'poison']`), une-os por vírgula e transforma as iniciais em maiúsculas (`Grass, Poison`).
* `for habilidade in pokemon.get("habilidades", []):`: Executa um laço de repetição percorrendo cada habilidade enviada pelo backend.
* `habilidades_html += f"<span class='badge'>..."`: Concatena dinamicamente cada habilidade dentro de uma tag `<span>` formatada com o design da nossa classe de pílula arredondada (`.badge`).
* `st.markdown(habilidades_html, unsafe_allow_html=True)`: Cospe o conjunto total de pílulas geradas lado a lado na interface web.
* `st.markdown("</div>", unsafe_allow_html=True)`: Fecha com segurança o escopo do bloco de visualização para assegurar a consistência do HTML interpretado pelo Streamlit.
