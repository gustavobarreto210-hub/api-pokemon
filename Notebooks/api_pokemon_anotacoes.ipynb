# Criando a API que consome a PokéAPI




## 1. O que preciso construir

Uma API com o seguinte endpoint:

```text
GET /pokedex/{nome-pokemon}
```

Exemplo:

```text
http://127.0.0.1:8000/pokemon/pikachu
```

Esse endpoint vai buscar informações do Pokémon na PokéAPI e retornar dados como:

- id
- nome
- altura
- peso
- tipos
- habilidades
- imagem


## 2. Criando a pasta do projeto

No terminal, execute:

```bash
mkdir api-pokemon
cd api-pokemon
```

A pasta `api-pokemon` será a pasta principal do projeto.


## 3. Criando o ambiente virtual

O ambiente virtual serve para isolar as bibliotecas desse projeto.

```bash
python -m venv venv
```

Ativando no Linux/Fedora:

```bash
source venv/bin/activate
```

Ativando no Windows:

```bash
venv\Scripts\activate
```


## 4. Instalando as dependências

É preciso instalar:

- `fastapi`: framework para criar a API
- `uvicorn`: servidor que executa a API
- `httpx`: biblioteca para fazer requisições HTTP para outra API

```bash
pip install fastapi uvicorn httpx
```


## 5. Criando o arquivo principal

Crie um arquivo chamado:

```text
main.py
```

Esse arquivo será o ponto de entrada da API.


````python
# Código do arquivo main.py

from fastapi import FastAPI, HTTPException
import httpx

app = FastAPI(title="Minha API Pokémon")

POKEAPI_URL = "https://pokeapi.co/api/v2"


@app.get("/")
def home():
    return {"mensagem": "API Pokémon funcionando!"}


@app.get("/pokemon/{nome}")
async def buscar_pokemon(nome: str):
    async with httpx.AsyncClient() as client:
        resposta = await client.get(f"{POKEAPI_URL}/pokemon/{nome.lower()}")

    if resposta.status_code == 404:
        raise HTTPException(status_code=404, detail="Pokémon não encontrado")

    if resposta.status_code != 200:
        raise HTTPException(status_code=500, detail="Erro ao consultar a PokéAPI")

    dados = resposta.json()

    return {
        "id": dados["id"],
        "nome": dados["name"],
        "altura": dados["height"],
        "peso": dados["weight"],
        "tipos": [tipo["type"]["name"] for tipo in dados["types"]],
        "habilidades": [
            habilidade["ability"]["name"]
            for habilidade in dados["abilities"]
        ],
        "imagem": dados["sprites"]["front_default"]
    }
````

## 6. Explicando o código

### Importações

```python
from fastapi import FastAPI, HTTPException
import httpx
```

- `FastAPI`: usado para criar a aplicação.
- `HTTPException`: usado para retornar erros HTTP, como erro 404.
- `httpx`: usado para fazer requisições para a PokéAPI.

### Criando a aplicação

```python
app = FastAPI(title="Minha API Pokémon")
```

Aqui criamos a aplicação FastAPI.

### URL base da PokéAPI

```python
POKEAPI_URL = "https://pokeapi.co/api/v2"
```

Essa variável guarda o endereço base da API externa que vamos consumir.


## 7. Rodando a API

No terminal, dentro da pasta do projeto, execute:

```bash
uvicorn main:app --reload
```

Explicando o comando:

- `uvicorn`: servidor que executa a API.
- `main`: nome do arquivo `main.py`.
- `app`: nome da variável que contém a aplicação FastAPI.
- `--reload`: reinicia automaticamente o servidor quando o código é alterado.


## 8. Acessando a documentação automática

Após rodar a API, acesse:

```text
http://127.0.0.1:8000/docs
```

Essa página é gerada automaticamente pelo FastAPI e permite testar os endpoints da API.
